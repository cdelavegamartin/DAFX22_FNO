```python
from WaveSolver2D import WaveSolver2D
import numpy as np
from numpy import sin, cos, conj, cumsum, real

waveSolver2D = WaveSolver2D()

xe = 0.2 # relative position
ye = 0.2 # 
dur = 0.01 # duration seconds
fe_xy = waveSolver2D.create_impulse(xe,ye)
# fe_xy = waveSolver2D.create_random_initial()
t, y, ybar, y_sp = waveSolver2D.solve(fe_xy,dur)
```

```python
from matplotlib import pyplot as plt 
plt.rcParams['text.usetex'] = True

# plot time signal
# y = real(y_x[20,:])
plt.plot(t,y)
plt.ylabel('$y**(x_\mathrm{o},t)$', fontsize = 14) 
plt.xlabel('Time $t$ in $[\mathrm{s}]$', fontsize=14)


# plot time space
plt.figure()
plt.imshow(np.transpose(real(y_sp[:,:,0])), interpolation='nearest')

plt.figure()
plt.imshow(np.transpose(real(y_sp[:,:,10])), interpolation='nearest')
```

```python
import IPython.display as ipd
# Sound output
y = real(y) 
y = y/max(y)
ipd.Audio(y, rate=waveSolver2D.Fs) # load a NumPy array
```
