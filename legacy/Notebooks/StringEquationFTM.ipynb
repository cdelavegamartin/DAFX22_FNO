```python
from StringSolver import StringSolver
import numpy as np
from numpy import sin, cos, conj, cumsum, real

stringSolver = StringSolver()

xe = 0.2 # relative position 
dur = 0.1 # duration seconds
fe_x = stringSolver.create_pluck(xe)
fe_x = stringSolver.create_random_initial()
t, y_x, y_defl_x = stringSolver.solve(fe_x,dur)
```

```python
from matplotlib import pyplot as plt 
plt.rcParams['text.usetex'] = True

# plot time signal
y = real(y_x[20,:])
plt.plot(t,y)
plt.ylabel('$y^(x_\mathrm{o},t)$', fontsize = 14) 
plt.xlabel('Time $t$ in $[\mathrm{s}]$', fontsize=14)

print(np.max(real(y_x)))
# plot time space
plt.figure(figsize = (10,20))
plt.imshow(np.transpose(real(y_x)), interpolation='nearest')
```

```python
import IPython.display as ipd
# Sound output
y = real(y) 
y = y/max(y)
ipd.Audio(y, rate=stringSolver.Fs) # load a NumPy array
```
